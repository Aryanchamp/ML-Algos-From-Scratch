# LinearRegression From Scratch 
 

#### Implementing using only pandas // comparing the results with sklearn LinearRegression model

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('tvmarketing.csv')
X = df.iloc[:,0].values.reshape(-1,1)
y = df.iloc[:,1].values.reshape(-1)
```

```python
w = 0.0
b = 0.0
learning_rate = 0.00001
epochs = 10000
n = len(y)
```

```python
for i in range(epochs):
    loss = 0
    db = 0
    dw = 0

    for j in range(n):
        y_pred = w * X[j] + b

        loss = loss + (y_pred - y[j]) ** 2

    total_loss = loss / (2 * n)


    for j in range(n):
        d_w = ((w * X[j] + b) - y[j]) * X[j]
        d_b = ((w * X[j] + b) - y[j])

        dw = dw + d_w
        db = db + d_b

    w = w - learning_rate * (dw/n)
    b = b - learning_rate * (db/n)


    if i % 100 == 0:
        print(f"Epoch {i}: Loss = {float(total_loss):.4f}, w = {float(w):.4f}, b = {float(b):.4f}")


print(f"\nFinal parameters -> w = {float(w):.3f}, b = {float(b):.3f}")

y_pred = w * X + b  ### calculating y_pred for all the values of x

ss_res = ((y - y_pred) ** 2).sum()       # residual sum of squares
ss_tot = ((y - y.mean()) ** 2).sum()     # total sum of squares
r2 = 1 - (ss_res / ss_tot)
print(f"R^2 Score: {r2:.4f}")
```

```python
plt.scatter(X, y, color='Blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Fitted line')
plt.xlabel('TV Advertisement Budget')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
